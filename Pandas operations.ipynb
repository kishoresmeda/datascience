## Renaming a column in Pandas

```python
ufo_df = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo_df.head()
```

```python
ufo_df.columns
```

### Method 1: Renaming individually

```python
ufo_df.rename(columns={'Colors Reported':'Colors_Reported','Shape Reported':'Shape_Reported'},inplace=True)
```

```python
ufo_df.columns
```

### Method 2: Renaming all columns to have _ in place of white spaces

```python
ufo_df.columns=ufo_df.columns.str.replace(' ','_')
ufo_df.columns
```

## > Deleting columns

```python
ufo_df2 = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo_df2.head()
```

### delete single column

```python
ufo_df2.drop('City', axis=1, inplace=True)
ufo_df2.head()
```

### delete multiple columns

```python
ufo_df2.drop(['State','Time'],axis=1,inplace=True)
ufo_df2.head()
```

### delete rows

```python
ufo_df2.drop([2,3],axis=0,inplace=True)
ufo_df2.head()
# 0th and 1st row selected for deletion. note axis=0 represents row and is optional since it's a default parameter to drop method
```

## Sorting a data frame

### sort a data frame by a column

```python
imdb_movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
imdb_movies.head(5)
```

```python
imdb_movies.sort_values('star_rating').head(5)
```

```python
imdb_movies.sort_values('duration').head(5)
```

```python
imdb_movies.sort_values('title').tail(20)
```

### sorting in descending order

```python
imdb_movies.sort_values('title',ascending=False).head(10)
#to apply any sorting to a data frame, add inplace=True argument
```

### Sorting pandas using multiple columns

```python
imdb_movies.sort_values(['content_rating','duration']).head(5)
# first it sorts using 'content_rating' and then applies sorting by duration to the resulting dataframe
# sorts in the order of arguments mentioned
```

### sorting a pandas series

```python
imdb_movies.title.sort_values()
```
